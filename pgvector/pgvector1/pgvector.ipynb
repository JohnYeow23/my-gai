````python
from langchain.document_loaders  import TextLoader
from langchain.text_splitter  import RecursiveCharacterTextSplitter
from langchain.embeddings  import OpenAIEmbeddings

from dotenv import load_dotenv

load_dotenv()
````

````python
loader = TextLoader('state_of_the_union.txt', encoding='utf-8')
documents = loader.load()
````

````python
# 1000 chararcters are being chosen, after creating one 80 characters will be taken from the prrevious creation an be included
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=80) 

# Split the text
texts = text_splitter.split_documents(documents)
````

# Why do we use a vector embeddings?

## Well this is going to cost you money at the end of the day if you have to keep sending your query into openai for embeddings and the documents are repeated with very minimal changes.
## Alot faster as compared to using an embeddings

````python
from langchain.vectorstores.pgvector import PGVector

CONNECTION_STRING = "postgresql+psycopg2://postgres:Thunder0981!@localhost:5432/vectordatabase"
COLLECTION_NAME = 'state_of_union_vectors'

db = PGVector.from_documents(
    embedding=embeddings, 
    documents=texts,
    collection_name=COLLECTION_NAME,
    connection_string=CONNECTION_STRING
)
````

````python
query = "What did the president say about Russia?"

similar = db.similarity_search_with_score(query, k=2) #The k finds the top two similar results

for doc in similar:
    print('-------------')
    print(doc[0].page_content)
    print(doc[1])
    print('-------------')
    for item in doc:
        print(item)
````

## Additional stuff to learn

If you’re using similarity_search_with_score, look at how your queries are written, specifically the operator: 
<br>
<=> for cosine similarity
<br>
<-> for Euclidean distance
<br>
<#> for inner product

````python
import psycopg2

# Connect to your PostgreSQL database
conn = psycopg2.connect("dbname=your_db user=your_user password=your_password")
cur = conn.cursor()

# Define your query vector as a list
query_vector = [0.1, 0.2, 0.3, 0.4]

# Convert the query vector to a PostgreSQL array format
query_vector_str = "ARRAY" + str(query_vector)

# Write the SQL query
sql = f"""
SELECT id, vector_column, (vector_column <-> {query_vector_str}) AS distance
FROM langchain_pg_embedding
ORDER BY distance
LIMIT 5;
"""

# Execute the query
cur.execute(sql)
results = cur.fetchall()

# Print results
for result in results:
    print(result)

# Close the connection
cur.close()
conn.close()
````
